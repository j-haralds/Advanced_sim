One of the best way to get a feeling for the statistical distributions we will encounter is to play around with them in a notebook. Here we will use the pdf's defined in the Python module 'scipy' (imported in the beginning of the notebook).

```python
# relevant modules necessary for this notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.stats import gamma, invgamma, t, norm, norminvgauss, mode
import seaborn as sns
sns.set_context("paper", font_scale=1.5)
sns.set_style("darkgrid")
sns.set_palette("deep")
sns.set(font='sans-serif')
%matplotlib inline
plt.rcParams['figure.dpi'] = 200
```

## The normal distribution

```python
def normal_distribution(mu,sigma2):
    return norm(loc=mu,scale=np.sqrt(sigma2)), [mu,sigma2]

th_var = np.linspace(-5,5,300)

N, pars = normal_distribution(mu=0,sigma2=1)
plt.plot(th_var,N.pdf(th_var),label=f'$\mu={pars[0]:.1f},\sigma={pars[1]:.1f}$');

N, pars = normal_distribution(mu=0,sigma2=2)
plt.plot(th_var,N.pdf(th_var),label=f'$\mu={pars[0]:.1f},\sigma={pars[1]:.1f}$');

N, pars = normal_distribution(mu=1,sigma2=1)
plt.plot(th_var,N.pdf(th_var),label=f'$\mu={pars[0]:.1f},\sigma={pars[1]:.1f}$');

plt.title(r'Normal distribution $\mathcal{N}(\theta|\mu,\sigma^2)$')
plt.xlabel(r'$\theta$');
plt.legend(facecolor='white');
```

## The gamma distribution

```python
def g_distribution(alpha,beta):
    return gamma(a=alpha,scale=beta), [alpha,beta]

s2_var = np.linspace(0,5,300)

G, pars = g_distribution(alpha=1,beta=1)
plt.plot(s2_var,G.pdf(s2_var),label=f'$\\alpha={pars[0]:.1f},\\beta={pars[1]:.1f}$');

G, pars = g_distribution(alpha=3,beta=1)
plt.plot(s2_var,G.pdf(s2_var),label=f'$\\alpha={pars[0]:.1f},\\beta={pars[1]:.1f}$');

G, pars = g_distribution(alpha=1,beta=2)
plt.plot(s2_var,G.pdf(s2_var),label=f'$\\alpha={pars[0]:.1f},\\beta={pars[1]:.1f}$');

G, pars = g_distribution(alpha=3,beta=2)
plt.plot(s2_var,G.pdf(s2_var),label=f'$\\alpha={pars[0]:.1f},\\beta={pars[1]:.1f}$');



plt.title(r'Gamma distribution $\mathcal{G}(\sigma^2|\alpha,\beta)$')
plt.xlabel(r'$\sigma^2$');
plt.legend(facecolor='white');
```

## The inverse-gamma distribution

```python
def ig_distribution(alpha,beta):
    return invgamma(a=alpha,scale=beta), [alpha,beta]
   
s2_var = np.linspace(0,5,300)

IG, pars = ig_distribution(alpha=1,beta=1)
plt.plot(s2_var,IG.pdf(s2_var),label=f'$\\alpha={pars[0]:.1f},\\beta={pars[1]:.1f}$');

IG, pars = ig_distribution(alpha=3,beta=1)
plt.plot(s2_var,IG.pdf(s2_var),label=f'$\\alpha={pars[0]:.1f},\\beta={pars[1]:.1f}$');

IG, pars = ig_distribution(alpha=1,beta=2)
plt.plot(s2_var,IG.pdf(s2_var),label=f'$\\alpha={pars[0]:.1f},\\beta={pars[1]:.1f}$');

plt.title(r'Inverse-gamma distribution $\mathcal{IG}(\sigma^2|\alpha,\beta)$')
plt.xlabel(r'$\sigma^2$');
plt.legend(facecolor='white');
```

## The normal-inverse-gamma distribution

```python
def nig(th,s2,mu0,Sigma0,alpha,beta):

    if alpha<=0:
        print('error alpha<=0')
        return 0
    if beta<=0:
        print('error beta<=0')
        return 0
    
    NIG = norm.pdf(th,loc=mu0,scale=np.sqrt(s2)*np.sqrt(Sigma0))*invgamma.pdf(s2,a=alpha,scale=beta) 
    return NIG   
```

```python
def nig_plot(mu,sigma,alpha,beta,plot_pars):

    ngrd   = plot_pars[0]   
    th_min = plot_pars[1]
    th_max = plot_pars[2]
    s2_min = plot_pars[3]
    s2_max = plot_pars[4]
    
    if s2_min <= 0:
        s2_min = 1e-6

    th = np.linspace(th_min,th_max,ngrd)
    s2 = np.linspace(s2_min,s2_max,ngrd)
    th_grid,s2_grid = np.meshgrid(th,s2)

    nig_pdf = np.zeros((ngrd,ngrd))
    for th_idx, th_val in enumerate(th):
        for s2_idx, s2_val in enumerate(s2):
            nig_pdf[s2_idx][th_idx] = nig(th_val,s2_val,mu,sigma,alpha,beta)
        
    fig1 = plt.figure(figsize=(5,5))
    contours = plt.contour(th_grid, s2_grid, nig_pdf, 5, colors='black')
    plt.imshow(nig_pdf, extent=[th_min,th_max,s2_min,s2_max], origin='lower',cmap='afmhot', alpha=0.9, aspect='auto');
    plt.colorbar();
    plt.title(r'$\mathcal{NIG}(\mu=%.1f,\Sigma=%.1f,\alpha=%.1f,\beta=%.1f)$'%(mu,sigma,alpha,beta))
    plt.xlabel(r'$\theta$');
    plt.ylabel(r'$\sigma^2$');
    plt.tight_layout();
    #plt.savefig('NIG_pdf.pdf');
    return fig1

alpha_test = 1
beta_test  = 1
mu_test    = 0
sigma_test = 2
nig_plot(mu_test,sigma_test,alpha_test,beta_test,plot_pars = [100,-4,+4,0,2.0]);
```
